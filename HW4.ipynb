```python
#sum each state
grouped_df = df.groupby('Jurisdiction of Occurrence')['COVID-19 (U071, Multiple Cause of Death)'].sum().reset_index()
grouped_df = grouped_df.drop([33,46]).reset_index()
grouped_df = grouped_df.drop(columns='index')
grouped_df = grouped_df.rename(columns={'COVID-19 (U071, Multiple Cause of Death)':'Total Deaths','Jurisdiction of Occurrence':'State'})
grouped_df
# prop_df = grouped_df
# prop_df
```

```python
pop2022 = state_pop.groupby(['STATE','NAME'])['POPESTIMATE2022'].sum()
pop2021 = state_pop.groupby(['STATE','NAME'])['POPESTIMATE2021'].sum()
pop2020 = state_pop.groupby(['STATE','NAME'])['POPESTIMATE2020'].sum()
avg_pop = (pop2022+pop2021+pop2020)/3
avg_pop = avg_pop.drop(0).reset_index().sort_values(by='NAME')
avg_pop = avg_pop.rename(columns={0:'Population','STATE':'id','NAME':'State'})
# avg_pop = avg_pop.drop(columns=['index'])
avg_pop
```
